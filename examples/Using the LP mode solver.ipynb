```python
from pyfiberamp.mode_solver import LPModeSolver, LPMode
```

```python
mode_solver = LPModeSolver(l_max=10, m_max=5)
all_modes = mode_solver.find_all_modes(core_radius=10e-6,
                          na=0.065,
                          wl=1064e-9)

for mode in all_modes:
    mode.plot_intensity()
    print(mode)
    
```
